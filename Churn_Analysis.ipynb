```python
def remove_outliers_iqr(df, column):
    # Convert to numeric, force errors to NaN
    df[column] = pd.to_numeric(df[column], errors='coerce')
    
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ✅ Function to annotate stacked bars
def annotate_stacked_bars(ax, textsize=12):
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            x, y = p.get_xy()
            ax.text(x + p.get_width() / 2,
                    y + height / 2,
                    f'{height:.1f}%',
                    ha='center',
                    va='center',
                    fontsize=textsize,
                    color='black')

# ✅ Function to plot stacked bar chart
def plot_stacked_bars(dataframe, title_, size_=(5, 5), rot_=0, legend_="lower right"):
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_,
        color=['skyblue', 'orange']
    )
    annotate_stacked_bars(ax, textsize=12)
    plt.legend(["Retention", "Churn"], loc=legend_)
    plt.ylabel("Company base (%)")
    plt.xlabel("")
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# ✅ Group by churn and count id
churn_counts = data_2.groupby('churn')['id'].count().reindex([0, 1], fill_value=0)

# ✅ Convert to percentage format
churn_df = pd.DataFrame({'Companies': churn_counts})
churn_percentage = churn_df / churn_df.sum() * 100

# ✅ Plot the chart
plot_stacked_bars(churn_percentage.transpose(), "Churning Status", (6, 5), legend_="lower right")
```
